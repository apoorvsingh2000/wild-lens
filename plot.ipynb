```python
import torch.nn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
losses = torch.load('../Best_Models/vit_full/model-losses.pth')
```

```python
losses
```

```python
sns.set(style='darkgrid')
```

```python
plt.figure(figsize=(10, 6))

plt.plot(range(losses['num_epochs']), losses['validation_losses'], label='Validation loss', color='red')
plt.plot(range(losses['num_epochs']), losses['train_losses'], label='Train loss', color='blue')
plt.title('Vision Transformers Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig('ViT-Full-loss_curves.png')

plt.show()
```
